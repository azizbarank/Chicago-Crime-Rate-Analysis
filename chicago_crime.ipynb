```python
# dropping columns with a majority of missing values
df = df.drop(axis=1, columns=['Ward',
                              'Community Area',
                              'Case Number',
                              'IUCR',
                              'Beat',
                              'District',
                              'FBI Code',
                              'Updated On'])
```

```python
df.info()
```

```python
df.Date = pd.to_datetime(df.Date, format='%m/%d/%Y %I:%M:%S %p')
```

```python
# Ensure 'Date' is set as the index and in datetime format
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Resample data yearly to get the count of crimes per year
crime_counts = df.resample('Y').size()

# Calculate the average crime rate across all years
average_crime_rate = crime_counts.mean()

# Create a detailed line plot
plt.figure(figsize=(14, 8))
plt.plot(crime_counts, label='Crimes Per Year', color='blue', linewidth=2, marker='o')
plt.axhline(y=average_crime_rate, color='red', linestyle='--', linewidth=1.5, label=f'Average Crime Rate: {average_crime_rate:.0f}')

# Add plot titles and labels
plt.title('Crimes Count Per Year (2001-2024)', fontsize=16)
plt.xlabel('Years', fontsize=14)
plt.ylabel('Number of Crimes', fontsize=14)

# Improve x-axis and y-axis ticks for better readability
plt.xticks(ticks=pd.date_range(start=crime_counts.index.min(), end=crime_counts.index.max(), freq='Y'), rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot for crime counts per year
plt.figure(figsize=(14, 8))
plt.scatter(crime_counts.index, crime_counts, color='green', label='Crimes Per Year', s=80, alpha=0.7)
plt.axhline(y=average_crime_rate, color='red', linestyle='--', linewidth=1.5, label=f'Average Crime Rate: {average_crime_rate:.0f}')

# Add plot titles and labels
plt.title('Crimes Count Scatter Plot Per Year (2001-2024)', fontsize=16)
plt.xlabel('Years', fontsize=14)
plt.ylabel('Number of Crimes', fontsize=14)

# Improve x-axis and y-axis ticks for better readability
plt.xticks(ticks=pd.date_range(start=crime_counts.index.min(), end=crime_counts.index.max(), freq='Y'), rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Group the data by year and get monthly counts to observe fluctuations within each year
df['Year'] = df.index.year  # Extract the year from the Date index

# Create a boxplot showing the distribution of monthly crime counts for each year
plt.figure(figsize=(18, 10))  # Adjust the figure size for better visualization
plt.boxplot(
    [df[df['Year'] == year].resample('M').size() for year in sorted(df['Year'].unique())],
    labels=sorted(df['Year'].unique()),
    patch_artist=True,  # To control box fill
    widths=0.7,  # Adjust the width of the boxes
    boxprops=dict(facecolor='none', color='blue'),  # Transparent boxes with blue outlines
    medianprops=dict(color='red', linewidth=2),  # Median line properties
    whiskerprops=dict(color='black', linestyle='--'),  # Whisker line properties
    capprops=dict(color='black'),  # Cap line properties
    flierprops=dict(marker='o', color='red', alpha=0.5)  # Outlier properties
)

# Add plot titles and labels
plt.title('Crime Count Distribution Per Year (Boxplot) (2001-2024)', fontsize=18)
plt.xlabel('Years', fontsize=14)
plt.ylabel('Number of Crimes Per Month', fontsize=14)

# Improve x-axis ticks for better readability
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.countplot(
    y='Location Description',
    data=df,
    order=df['Location Description'].value_counts().iloc[:10].index,
    palette='viridis'  # Use a colorful palette like 'viridis', 'plasma', 'coolwarm', etc.
)

# Add titles and labels
plt.title('Top 10 Crime Locations', fontsize=18)
plt.xlabel('Number of Crimes', fontsize=14)
plt.ylabel('Location Description', fontsize=14)

# Improve layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# Plotting categorical count plot for crime by area with a colorful palette
sns.countplot(
    y='Block',
    data=df,
    orient='h',
    order=df['Block'].value_counts()[:10].index,
    palette='viridis'  # Use a vibrant palette like 'coolwarm', 'viridis', or 'Set2'
)

# Setting x-axis label and figure title
plt.xlabel('Volume of Crime in an Area', fontsize=12)
plt.title('Chicago - Volume of Crime', fontsize=16)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
